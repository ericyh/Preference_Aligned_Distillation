```python
class FashionCLIPImageEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        full_clip_model = CLIPModel.from_pretrained("patrickjohncyh/fashion-clip")
        self.vision_encoder = full_clip_model.vision_model
        self.projection = full_clip_model.visual_projection
        del full_clip_model

    def forward(self, pixel_values):
        vision_outputs = self.vision_encoder(pixel_values=pixel_values)
        pooled_output = vision_outputs.pooler_output
        x = self.projection(pooled_output)
        return x
```

```python
model = FashionCLIPImageEncoder().to("cuda")
```
