```python
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
df = pd.read_parquet("train-00000-of-00001.parquet")
```

```python
data = df["persona"].tolist()
```

```python
random.shuffle(data)
```

```python
test_data = data[:1000]
validation_data = data[1000:2000]
train_data = data[2000:7000]
```

```python
len(test_data), len(train_data), len(validation_data)
```

```python
with open('test_personas.json', 'w') as f:
    json.dump(test_data, f)
with open('train_personas.json', 'w') as f:
    json.dump(train_data, f)
with open('validation_personas.json', 'w') as f:
    json.dump(validation_data, f)
```

```python
with open('test_personas.json', 'r') as f:
    test_personas = json.load(f)
```
