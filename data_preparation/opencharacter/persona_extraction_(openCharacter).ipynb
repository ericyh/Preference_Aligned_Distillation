```python
import pandas as pd
import numpy as np
import json
import random
import openai
```

```python
with open('synthetic_character.jsonl', 'r') as file:
    all_data = [json.loads(line)["persona"] for line in file]
```

```python
data = []
```

```python
from langdetect import detect, DetectorFactory
DetectorFactory.seed = 0  # for consistent results

print(len(all_data))
english_only = []
for s in all_data:
    try:
        if detect(s) == "en":
            data.append(s)
    except:
        pass  # skip if detection fails

print(len(data))
```

```python
random.shuffle(data)
```

```python
test_personas = data[:1000]
validation_personas = data[1000:2000]
train_personas = data[2000:]
```

```python
len(test_personas), len(train_personas), len(validation_personas)
```

```python
with open('test_personas.json', 'w') as f:
    json.dump(test_personas, f)
with open('train_personas.json', 'w') as f:
    json.dump(train_personas, f)
with open('validation_personas.json', 'w') as f:
    json.dump(validation_personas, f)
```
